```python
#¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_vuelos["Year"] = df_vuelos["Year"].astype(int)
df_vuelos["Month"] = df_vuelos["Month"].astype(int)
```

```python
df_vuelos["Date"] = pd.to_datetime(
    df_vuelos["Year"].astype(str) + "-" + df_vuelos["Month"].astype(str) + "-01"
)
```

```python
df_vuelos["Month_Name"] = df_vuelos["Date"].dt.month_name(locale="Spanish")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(
    data=df_vuelos,
    x="Month_Name",
    weights="Flights Booked",
    color="skyblue",
    alpha=0.7
)

plt.title("Distribución de vuelos reservados por mes")
plt.xlabel("Mes")
plt.ylabel("Cantidad de vuelos reservados")
plt.xticks(rotation=45)
plt.show()
```

```python
# ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente? 
# con tendencias o correlaciones
```

```python
sns.scatterplot(
    data=df_vuelos,
    x="Distance",             # eje X
    y="Points Accumulated",   # eje Y
   marker='d', color = 'red', s=10
)
plt.title("Relación entre distancia y puntos acumulados por cliente")
plt.xlabel("Distancia total de vuelos")
plt.ylabel("Puntos acumulados")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)  # leyenda fuera del gráfico
plt.show()
```

```python
### relación positiva y lineal. PArece haber cuatro clases de vuelos. No hay dispersión aleatoria. 
```

```python
#¿Cuál es la distribución de los clientes por provincia o estado?
clientes_por_provincia = df_clientes['Province'].value_counts().reset_index()
clientes_por_provincia.columns = ['Province', 'Num_Clientes']
clientes_por_provincia
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=clientes_por_provincia, x='Province', y='Num_Clientes')
plt.xticks(rotation=45, ha='right')
plt.title('Distribución de clientes por provincia')
plt.ylabel('Número de clientes')
plt.xlabel('Provincia')
plt.show()
```

```python
#¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?
salario_por_educacion = df_clientes.groupby('Education')['Salary'].mean().reset_index()
salario_por_educacion
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=salario_por_educacion, x='Education', y='Salary')
plt.xticks(rotation=45, ha='right')
plt.title('Salario promedio por nivel educativo')
plt.ylabel('Salario promedio')
plt.xlabel('Nivel educativo')
plt.tight_layout()
plt.show()
```

```python
#¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad? 
tarjetas_counts = df_clientes['Loyalty Card'].value_counts()
tarjetas_proporcion = tarjetas_counts / tarjetas_counts.sum() * 100
tarjetas_proporcion = tarjetas_proporcion.reset_index()
tarjetas_proporcion.columns = ['Loyalty Card', 'Proporcion (%)']
tarjetas_proporcion
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=tarjetas_proporcion, x='Loyalty Card', y='Proporcion (%)')
plt.title('Proporción de clientes por tipo de tarjeta de fidelidad')
plt.ylabel('Proporción (%)')
plt.xlabel('Tipo de tarjeta')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ¿Cómo se distribuyen los clientes según su estado civil y género?
estado_genero = pd.crosstab(df_clientes['Marital Status'], df_clientes['Gender'])
estado_genero
```

```python
estado_genero.plot(kind='bar', figsize=(10,6), color=['blue','red'])
plt.title('Distribución de clientes por estado civil y género')
plt.ylabel('Número de clientes')
plt.xlabel('Estado civil')
plt.xticks(rotation=0)
plt.legend(title='Género')
plt.tight_layout()
plt.show()
```
