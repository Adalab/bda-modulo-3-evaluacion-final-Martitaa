##DATA CLIENTES

```python
df_clientes = pd.read_csv("Customer Loyalty History.csv", index_col=0)
df_clientes
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_clientes.shape
```

```python
df_clientes.columns
```

```python
df_clientes.info()
```

```python
df_clientes.isnull().sum()
```

```python
#el salary lo cambiamos por la media de salarios los nulos 
df_clientes.describe().T
```

```python
df_clientes["Salary"].isnull().sum()
```

```python
mean_salary= df_clientes["Salary"].mean()
```

```python
df_clientes["Salary"].fillna(mean_salary, inplace=True)
```

```python
#comrpobamos si ahora la variable salary no tiene ningun valor nulo
df_clientes["Salary"].isnull().sum()
```

```python
#ahora cuatro variables quiero que sean cambiadas a datetime
cols = ["Enrollment Year", "Enrollment Month", "Cancellation Year", "Cancellation Month"]
for col in cols:
    df_clientes[col]=pd.to_datetime(df_clientes[col])
#comprobamos para ver si se ha cambiado
df_clientes.info()
```

```python
columnas_desconocido = ["Cancellation Year","Cancellation Month" ]
for columna in columnas_desconocido:
    df_clientes[columna] =df_clientes[columna].fillna("desconocido")
    print(f"Columna '{columna}: Nulos remplazados por 'desconocido'")
```

```python
df_clientes.info()
```

##UNIR los dos conjuntos de datos de la forma más eficiente

```python
df_vuelos.info()
```

```python
df_unido = pd.merge(
    df_vuelos,          # información mensual de vuelos
    df_clientes,        # perfil del cliente
    on="Loyalty Number",# columna clave común
    how="left"          # left join para mantener todos los vuelos aunque falte info de perfil
)
```
